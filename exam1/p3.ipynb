```python
import pandas as pd
import matplotlib.pyplot as plt

features = ["Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids",
           "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines",
           "Proline "]

data = pd.read_csv("wine.csv")
data = data.dropna()

for feature in features:
    plt.figure(figsize=(6, 3))
    classes = data['Class'].unique()
    for cls in classes:
        subset = data[data['Class'] == cls]
        plt.scatter(subset[feature], [cls] * len(subset), label=cls, alpha=0.7)

    # Adding labels and title
    plt.xlabel(feature)
    plt.ylabel('Class Label')
    plt.title(f'Scatter Plot of {feature} vs Class Label')
    plt.legend()

    # Display the plot
    plt.show()

#scroll to after plots for rest of question
```

It appears that Flavanoids has the best predictive power for predicting class based upon the class vs feature scatter plots. I say this because this model has the most concentrated groups for each class (ie, low variance), and also because there is the most clear division between each class.

```python
import matplotlib.pyplot as plt

print("Scatter plot of features 7 and 8 - Flavonoids and Nonflavanoid phenols")

# Extract features
X = data[['Flavanoids', 'Nonflavanoid phenols']].values

#class transistions:
# 0 - 58 = class 1
# 59 - 129 = class 2
# 130 - end = class 3

# Scatter plot for different classes, divide by when class ends
plt.scatter(X[:59, 0], X[:59, 1], color='red', marker='o', label='Class 1')
plt.scatter(X[59:130, 0], X[59:130, 1], color='blue', marker='s', label='Class 2')
plt.scatter(X[130:, 0], X[130:, 1], color='green', marker='x', label='Class 3')

plt.xlabel("Flavanoids")
plt.ylabel("Nonflavanoid phenols")
plt.legend()
plt.tight_layout()
plt.show()

```
